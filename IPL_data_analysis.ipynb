# 🏏 IPL Data Analysis — Matplotlib & Seaborn
### A data visualization project exploring IPL match trends, team performance, and player stats.
**Dataset:** Synthetically generated IPL-style data (2008–2023)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="darkgrid")
print("Libraries loaded successfully!")
```

```python
np.random.seed(42)

teams = ['Mumbai Indians', 'Chennai Super Kings', 'Royal Challengers Bangalore',
         'Kolkata Knight Riders', 'Delhi Capitals', 'Rajasthan Royals',
         'Sunrisers Hyderabad', 'Punjab Kings']

players = ['Virat Kohli', 'Rohit Sharma', 'MS Dhoni', 'AB de Villiers',
           'David Warner', 'KL Rahul', 'Jasprit Bumrah', 'Rashid Khan']

n = 500

df = pd.DataFrame({
    'match_id': range(1, n+1),
    'season': np.random.randint(2008, 2024, n),
    'team1': np.random.choice(teams, n),
    'team2': np.random.choice(teams, n),
    'winner': np.random.choice(teams, n),
    'toss_winner': np.random.choice(teams, n),
    'toss_decision': np.random.choice(['bat', 'field'], n),
    'top_scorer': np.random.choice(players, n),
    'runs_scored': np.random.randint(20, 120, n),
    'wickets': np.random.randint(0, 5, n),
    'margin': np.random.randint(1, 100, n),
    'venue': np.random.choice(['Wankhede', 'Eden Gardens', 'Chepauk',
                                'Chinnaswamy', 'Feroz Shah Kotla'], n)
})

print(df.head())
```

## 🏆 Plot 1 — Team Win Count
Which IPL team has won the most matches?

```python
plt.figure(figsize=(10, 6))
win_counts = df['winner'].value_counts()
colors = sns.color_palette('Set2', len(win_counts))

plt.barh(win_counts.index, win_counts.values, color=colors)
plt.title('🏆 IPL Team Win Count', fontsize=16, fontweight='bold')
plt.xlabel('Number of Wins', fontsize=12)
plt.ylabel('Team', fontsize=12)
plt.tight_layout()
plt.show()
```

## 🏏 Plot 2 — Toss Decision Analysis
Do teams prefer to bat or field after winning the toss?

```python
plt.figure(figsize=(8, 6))
toss_counts = df['toss_decision'].value_counts()
colors = ['#f4a261', '#2a9d8f']

plt.pie(toss_counts.values, labels=toss_counts.index,
        colors=colors, autopct='%1.1f%%',
        startangle=90, explode=[0.05, 0.05],
        textprops={'fontsize': 13})
plt.title('🏏 Toss Decision — Bat vs Field', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()
```

## 🌟 Plot 3 — Top Scorers
Which players scored the most runs across all matches?

```python
plt.figure(figsize=(10, 6))
top_scorers = df.groupby('top_scorer')['runs_scored'].sum().sort_values(ascending=False)
colors = sns.color_palette('rocket', len(top_scorers))

sns.barplot(x=top_scorers.values, y=top_scorers.index, palette='rocket')
plt.title('🌟 Total Runs by Player', fontsize=16, fontweight='bold')
plt.xlabel('Total Runs Scored', fontsize=12)
plt.ylabel('Player', fontsize=12)
plt.tight_layout()
plt.show()
```

## 📅 Plot 4 — Matches Per Season
How many matches were played each IPL season?

```python
plt.figure(figsize=(12, 6))
matches_per_season = df['season'].value_counts().sort_index()

plt.plot(matches_per_season.index, matches_per_season.values,
         color='#e63946', linewidth=2.5, marker='o', markersize=7)
plt.fill_between(matches_per_season.index, matches_per_season.values,
                 alpha=0.2, color='#e63946')
plt.title('📅 Number of Matches Per IPL Season', fontsize=16, fontweight='bold')
plt.xlabel('Season', fontsize=12)
plt.ylabel('Number of Matches', fontsize=12)
plt.xticks(matches_per_season.index, rotation=45)
plt.tight_layout()
plt.show()
```

## 🔥 Plot 5 — Venue vs Season Heatmap
Which venues hosted the most matches each season?

```python
plt.figure(figsize=(12, 6))
heatmap_data = df.groupby(['venue', 'season']).size().unstack(fill_value=0)

sns.heatmap(heatmap_data, annot=True, fmt='d',
            cmap='YlOrRd', linewidths=0.5,
            cbar_kws={'label': 'Number of Matches'})
plt.title('🔥 Matches Hosted by Venue per Season', fontsize=16, fontweight='bold')
plt.xlabel('Season', fontsize=12)
plt.ylabel('Venue', fontsize=12)
plt.tight_layout()
plt.show()
```
